```python
Dates=NIG.index[0:-1:10]
NCC=NIG['No. of confirmed cases'][0:-1:10]
NCD=NIG['No. of confirmed deaths'][0:-1:10]
ltv.plot(Dates,NCC,"o--r",ms=2,label="Infectados")
ltv.plot(Dates,NCD,"o-k",ms=4, label="Muertes")
ltv.grid()
ltv.title("Avance del ebola en Nigeria")
ltv.ylabel("Casos confirmados")
ltv.xlabel("Periodo de tiempo")
ltv.xticks(rotation=90, fontsize=7)
ltv.legend()
ltv.show()
```
