```python
Tarea : Reto 2. Convertidor de Unidades.

Integrantes: Arellano Sandoval Gustavo Adolfo (AS)

Nombre del programa : computation . ipynb


# Función para convertir unidades de longitud
def convertir_longitud():
    print("Opciones de conversión de longitud:")
    print("1. Metros a Pies")
    print("2. Pies a Metros")
    print("3. Kilómetros a Millas")
    print("4. Millas a Kilómetros")

    opcion = int(input("Seleccione una opción (1/2/3/4): "))

    if opcion == 1:
        metros = float(input("Ingrese la longitud en metros: "))
        pies = metros * 3.28084
        print(f"{metros} metros son aproximadamente {pies} pies.")
    elif opcion == 2:
        pies = float(input("Ingrese la longitud en pies: "))
        metros = pies / 3.28084
        print(f"{pies} pies son aproximadamente {metros} metros.")
    elif opcion == 3:
        km = float(input("Ingrese la distancia en kilómetros: "))
        millas = km * 0.621371
        print(f"{km} kilómetros son aproximadamente {millas} millas.")
    elif opcion == 4:
        millas = float(input("Ingrese la distancia en millas: "))
        km = millas / 0.621371
        print(f"{millas} millas son aproximadamente {km} kilómetros.")
    else:
        print("Opción no válida")

# Función para convertir unidades de masa
def convertir_masa():
    print("Opciones de conversión de masa:")
    print("1. Kilogramos a Libras")
    print("2. Libras a Kilogramos")
    print("3. Gramos a Onzas")
    print("4. Onzas a Gramos")

    opcion = int(input("Seleccione una opción (1/2/3/4): "))

    if opcion == 1:
        kg = float(input("Ingrese la masa en kilogramos: "))
        libras = kg * 2.20462
        print(f"{kg} kilogramos son aproximadamente {libras} libras.")
    elif opcion == 2:
        libras = float(input("Ingrese la masa en libras: "))
        kg = libras / 2.20462
        print(f"{libras} libras son aproximadamente {kg} kilogramos.")
    elif opcion == 3:
        gramos = float(input("Ingrese la masa en gramos: "))
        onzas = gramos * 0.035274
        print(f"{gramos} gramos son aproximadamente {onzas} onzas.")
    elif opcion == 4:
        onzas = float(input("Ingrese la masa en onzas: "))
        gramos = onzas / 0.035274
        print(f"{onzas} onzas son aproximadamente {gramos} gramos.")
    else:
        print("Opción no válida")

# Función para convertir unidades de tiempo
def convertir_tiempo():
    print("Opciones de conversión de tiempo:")
    print("1. Horas a Segundos")
    print("2. Segundos a Horas")
    print("3. Minutos a Días")
    print("4. Días a Minutos")

    opcion = int(input("Seleccione una opción (1/2/3/4): "))

    if opcion == 1:
        horas = float(input("Ingrese el tiempo en horas: "))
        segundos = horas * 3600
        print(f"{horas} horas son aproximadamente {segundos} segundos.")
    elif opcion == 2:
        segundos = float(input("Ingrese el tiempo en segundos: "))
        horas = segundos / 3600
        print(f"{segundos} segundos son aproximadamente {horas} horas.")
    elif opcion == 3:
        minutos = float(input("Ingrese el tiempo en minutos: "))
        dias = minutos / 1440
        print(f"{minutos} minutos son aproximadamente {dias} días.")
    elif opcion == 4:
        dias = float(input("Ingrese el tiempo en días: "))
        minutos = dias * 1440
        print(f"{dias} días son aproximadamente {minutos} minutos.")
    else:
        print("Opción no válida")

# Función principal
def main():
    while True:
        print("Menú Principal:")
        print("1. Convertir unidades de longitud")
        print("2. Convertir unidades de masa")
        print("3. Convertir unidades de tiempo")
        print("4. Salir del programa")

        opcion = int(input("Seleccione una opción (1/2/3/4): "))

        if opcion == 1:
            convertir_longitud()
        elif opcion == 2:
            convertir_masa()
        elif opcion == 3:
            convertir_tiempo()
        elif opcion == 4:
            print("¡Adiós!")
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción válida.")

if __name__ == "__main__":
    main()
```
