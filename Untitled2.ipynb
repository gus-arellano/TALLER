```python
import matplotlib.pyplot as plt
import numpy as np

# Crear un rango de valores para el eje x
x = np.linspace(0, 2 * np.pi, 100)  # Rango de 0 a 2*pi

# Calcular los valores de seno y coseno para cada valor en x
y_seno = np.sin(x)
y_coseno = np.cos(x)

# Crear el grfico
plt.figure(figsize=(14, 5))  # Establecer el tamao de la figura

# Graficar las funciones seno y coseno
plt.plot(x, y_seno, label='Seno', color='blue')
plt.plot(x, y_coseno, label='Coseno', color='red')

# Agregar una cuadrcula
plt.grid(axis='y', linestyle='--', color = 'green')
plt.grid(axis='x', linestyle='--', color = 'green')

# Etiquetas de los ejes
plt.xlabel('ngulo (radianes)')
plt.ylabel('Valor')

# Ttulo del grfico
plt.title('Grfico de Seno y Coseno')

# Leyenda
plt.legend()

# Mostrar el grfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def graficacin(y):
    # Crear un rango de valores para el eje x
    x = np.linspace(0, 2 * np.pi, 100)  # Rango de 0 a 2*pi

    # Calcular los valores de seno y coseno para cada valor en x
    y_seno = np.sin(x) + y
    y_coseno = np.cos(x) + y

    # Crear el grfico
    plt.figure(figsize=(14, 5))  # Establecer el tamao de la figura

    # Graficar las funciones seno y coseno
    plt.plot(x, y_seno, label=f'Seno (y={y})', color='blue')
    plt.plot(x, y_coseno, label=f'Coseno (y={y})', color='red')

    # Agregar una cuadrcula
    plt.grid(axis='y', linestyle='--', color = 'green')
    plt.grid(axis='x', linestyle='--', color = 'green')

    # Etiquetas de los ejes
    plt.xlabel('ngulo (radianes)')
    plt.ylabel('Valor')

    # Ttulo del grfico
    plt.title('Grfico de Seno y Coseno')

    # Leyenda
    plt.legend()

    # Mostrar el grfico
    plt.show()

# Llamamos a la funcin graficacin con un valor de y, por ejemplo, 0.5
graficacin(0.5)
```
