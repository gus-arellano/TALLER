```python
def es_primo(numero):
    """
    Verifica si un número es primo.
    Parameters:
    - numero (int): El número a verificar.
    Returns:
    - bool: True si el número es primo, False si no lo es.
    """
    # Asegurarse de que el número sea mayor que 1
    while numero <= 1:
        print("Error en el Número. El número debe ser mayor que 1.")
        numero = int(input("Ingrese un número válido: "))
    # Verificación de primalidad
    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False  # El número no es primo
    return True  # El número es primo
# Prueba
numero_usuario = int(input("Ingrese un número entero mayor que 1: "))
resultado = es_primo(numero_usuario)
# Resultado
print(f"¿{numero_usuario} es un número primo? {resultado}")
```

```python
import random
import matplotlib.pyplot as plt
# Calcular el promedio
def calcular_promedio(numeros):
    return sum(numeros) / len(numeros)
# Calcular la varianza
def calcular_varianza(numeros, promedio):
    return sum((x - promedio) ** 2 for x in numeros) / len(numeros)
# Calcular la desviación estándar
def calcular_desviacion_estandar(varianza):
    return varianza ** 0.5
# Arreglo de números enteros entre 0 y 150
numeros = [random.randint(0, 150) for _ in range(100)]
# Calcular promedio
promedio = calcular_promedio(numeros)
# Calcular varianza
varianza = calcular_varianza(numeros, promedio)
# Calcular desviación estándar
desviacion_estandar = calcular_desviacion_estandar(varianza)
# Visualizar datos en un histograma
plt.hist(numeros, bins=15, color='blue', edgecolor='black')
plt.title('Histograma de Números Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
# Calcular la suma de números pares e impares
suma_pares = sum(x for x in numeros if x % 2 == 0)
suma_impares = sum(x for x in numeros if x % 2 != 0)
# Calcular los múltiplos de 2, 3 y 5
multiplos_2 = [x for x in numeros if x % 2 == 0]
multiplos_3 = [x for x in numeros if x % 3 == 0]
multiplos_5 = [x for x in numeros if x % 5 == 0]
# Imprimir resultados
print("Promedio:", promedio)
print("Varianza:", varianza)
print("Desviación Estándar:", desviacion_estandar)
print("Suma de números pares:", suma_pares)
print("Suma de números impares:", suma_impares)
print("Múltiplos de 2:", multiplos_2)
print("Múltiplos de 3:", multiplos_3)
print("Múltiplos de 5:", multiplos_5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Función que representa la solución general
def solucion_general(x, C):
    return (x - 1) + C * np.exp(-x)
# Rango de valores para x
x_values = np.linspace(-2, 5, 100)
# Valores específicos de C para cada caso
C_positivo = 2
C_cero = 0
C_negativo = -2
# Calcular y para cada conjunto de valores de x y C
y_positivo = solucion_general(x_values, C_positivo)
y_cero = solucion_general(x_values, C_cero)
y_negativo = solucion_general(x_values, C_negativo)
# Graficar las curvas para cada caso
plt.plot(x_values, y_positivo, label='C > 0', color='blue')
plt.plot(x_values, y_cero, label='C = 0', color='green')
plt.plot(x_values, y_negativo, label='C < 0', color='red')
# Etiquetas y título del gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Familia de Curvas de la Solución de la Ecuación Diferencial')
plt.legend()  # Mostrar leyenda
# Marcar los puntos de intersección
plt.scatter([1], [0], color='black')  # Punto de intersección para C = 0
# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Definir la función
def funcion(x):
    return 3*x + 7*x**2 - 4
# Rango de valores de x con más de 1000 muestras
x_values = np.linspace(-10, 10, 1000)
# Valores de y correspondientes
y_values = funcion(x_values)
# Graficar la función con línea continua
plt.plot(x_values, y_values, label='Función: $3x + 7x^2 - 4$', color='blue')
# Graficar asíntotas horizontales y verticales
plt.axhline(0, linestyle='--', color='red', label='Asíntota Horizontal')
plt.axvline(0, linestyle='--', color='green', label='Asíntota Vertical')
# Establecer límites en el eje y para evitar valores muy altos o bajos cerca de las asíntotas verticales
plt.ylim(-100, 100)
# Añadir leyenda
plt.legend()
# Etiquetas y título del gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de la Función con Asíntotas')
# Mostrar el gráfico
plt.grid(True)
plt.show()
```
